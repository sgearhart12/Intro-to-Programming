```python
from bs4 import BeautifulSoup 
import requests
from urllib.parse import urljoin

#url that I will be scraping
#set the url to a variable
#because BeautifulSoup can only parse html, I'm going to look at a specific webpage with data (stroke clinical trials)
url='https://clinicaltrials.gov/search?cond=stroke'

#requests
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
print(soup)
print(soup.prettify())
```

```python
#I'd like to extract study titles
#titles stored in a list
titles=soup.select("a.hit-card-title.usa-card__heading")
title_results=[]
for t in titles:
    title=t.get_text(" ",strip=True)
    href=t.get("href"," ")
    full_url=urljoin(url,href)
    title_results.append({"title": title, "url": full_url})

for r in title_results:
    print(f"- {r['title']}\n  {r['url']}")

#troubleshooting
print(response.status_code)
print(len(response.text))
print(response.text[:500])

#unfortunately, I cannot parse this webpage with beautifulsoup since it contains more complext rendering. 
```

```python
#I'll use this example site to demonstrate scraping. 
url='https://realpython.github.io/fake-jobs/'
response=requests.get(url)
soup=BeautifulSoup(response.text, "html.parser")
#aim to extract job titles
titles=soup.find_all("h2", class_="title is-5")
for t in titles[:5]:
    print(t.get_text(strip=True))

#scraped titles successfully!



```
